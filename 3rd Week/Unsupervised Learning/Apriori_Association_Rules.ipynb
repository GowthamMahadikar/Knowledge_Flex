```python

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

## 1. Dataset

```python

dataset = [
    ['Milk', 'Bread', 'Eggs'],
    ['Milk', 'Bread'],
    ['Milk', 'Cheese'],
    ['Milk', 'Bread', 'Eggs'],
    ['Bread', 'Eggs']
]

dataset
```

## 2. One-Hot Encoding
Apriori requires data in a boolean (0/1) format.

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```


## 3. Frequent Itemset Generation
- **min_support = 0.6**


```python

frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```


## 4. Association Rules
- **metric = confidence**
- **min_threshold = 0.7**


```python

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.7
)

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```


## 5. Interpretation
- **Support**: How frequently an itemset appears  
- **Confidence**: How often the rule is correct  
- **Lift**: Strength of the rule (>1 means strong association)
