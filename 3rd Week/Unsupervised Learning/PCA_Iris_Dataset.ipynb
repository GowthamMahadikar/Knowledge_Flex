# Principal Component Analysis (PCA) â€“ Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

## 1. Load Dataset

```python
iris = load_iris()
X = iris.data
y = iris.target
X[:5]
```

## 2. Feature Scaling
Scaling is critical for PCA since it is variance-based.

```python
X_scaled = StandardScaler().fit_transform(X)
X_scaled[:5]
```

## 3. Apply PCA
Reducing from 4 dimensions to 2 principal components.


```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca[:5]
```

## 4. Explained Variance

```python

print(f"Explained Variance Ratio: {pca.explained_variance_ratio_}")
print(f"Total Variance Captured: {pca.explained_variance_ratio_.sum():.2f}")
```

## 5. Visualization
Each color represents a different Iris species.


```python
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(iris.target_names):
    plt.scatter(
        X_pca[y == i, 0],
        X_pca[y == i, 1],
        label=target_name
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.show()
```
